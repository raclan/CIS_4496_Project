```python
df_a.drop(['product_code', 'prod_name', 'product_type_name', 'graphical_appearance_name', 'colour_group_name', 
        'perceived_colour_value_id', 'perceived_colour_value_name','perceived_colour_master_id', 
        'perceived_colour_master_name', 'department_name', 'index_name', 'index_group_no', 'index_group_name',
        'section_name', 'garment_group_name', 'detail_desc'], inplace=True, axis=1)
```

```python
enc_df_a = df_a.copy().drop(df_a.columns[1:],axis=1)
enc = OneHotEncoder()

for col in df_a.columns[1:]:
    df_a[col]=df_a[col].astype('category')
    df_a[col]=df_a[col].cat.codes

    enc_temp = pd.DataFrame(enc.fit_transform(df_a[[col]]).toarray())

    enc_df_a = pd.concat([enc_df_a, enc_temp.reindex(enc_df_a.index)], axis=1)
```
