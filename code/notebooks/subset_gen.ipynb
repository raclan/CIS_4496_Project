```python
import polars as pl
import numpy as np
import random
```

```python
df_t = pl.read_csv('~/CIS_4496_Project/data/raw/transactions_train.csv')
df_t.head()
```

```python
df_t.height
```

```python
df_t_u = df_t.unique(subset=["customer_id"])
```

```python
customer_subset = df_t_u.sample(frac=0.01, seed=9).get_column("customer_id")
```

```python
customer_subset
```

```python
transactions_subset = pl.DataFrame(schema=df_t.schema)

for c in customer_subset:
    transactions_subset = transactions_subset.vstack(df_t.filter(pl.col("customer_id") == c))
transactions_subset
```
