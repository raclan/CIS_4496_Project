```python
df_a_temp = df_a.copy().drop(['product_code', 'prod_name', 'product_type_name', 'graphical_appearance_name', 'colour_group_name', 
        'perceived_colour_value_id', 'perceived_colour_value_name','perceived_colour_master_id', 
        'perceived_colour_master_name', 'department_name', 'index_name', 'index_group_no', 'index_group_name',
        'section_name', 'garment_group_name', 'detail_desc'], axis=1)
df_a_temp
```

```python
enc_df_a = df_a.copy().drop(df_a.columns[1:],axis=1)
enc = OneHotEncoder()

for col in df_a_temp.columns[1:]:
    df_a_temp[col]=df_a_temp[col].astype('category')
    df_a_temp[col]=df_a_temp[col].cat.codes

    enc_temp = pd.DataFrame(enc.fit_transform(df_a_temp[[col]]).toarray())

    enc_df_a = pd.concat([enc_df_a, enc_temp.reindex(enc_df_a.index)], axis=1)
```

```python
df_a_vectors = df_a.copy().drop(df_a.columns[1:],axis=1)
one_hot_vector = []

for i in range(len(enc_df_a)):
    one_hot_vector.append(np.array(enc_df_a.iloc[i][1:]))

df_a_vectors['one_hot_vector'] = one_hot_vector
df_a_vectors
```

```python
def csim(vector1, vector2):
    return np.dot(vector1, vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2))

def similarities(vector):
    temp = []
    for row in range(len(df_a_vectors)):
        temp.append([df_a_vectors['article_id'][row], csim(vector, df_a_vectors['one_hot_vector'][row])])
    return temp
```

```python
def tensor_csim(row):
    temp = torch.empty(0,1)
    to_return = torch.empty(len(df_a_vectors['one_hot_vector']), 0)
    for r in df_a_vectors['one_hot_vector']:
        temp = torch.cat( (temp, torch.tensor([[csim(row, r)]])), 0)
    
    return torch.cat( (to_return, temp), 1)

def tensor_sim():
    to_return = torch.empty(len(df_a_vectors['one_hot_vector']), 0)
    for row in df_a_vectors['one_hot_vector']:
        torch.cat( (to_return, tensor_csim(row)), 1)
    return to_return
```
